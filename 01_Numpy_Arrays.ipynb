```python
import numpy as np

# Array
arr1 = np.array([1, 2, 3, 4, 5])
print("\nArray:", arr1)
print("Elements type:", arr1.dtype)

# Array of Zeros
arr2 = np.zeros(5)
print("\nArray of zeros:", arr2)
print("Elements type:", arr2.dtype)

# Array of Ones
arr3 = np.ones(5)
print("\nArray of ones:", arr3)
print("Elements type:", arr3.dtype)

# Array of Random Numbers
arr4 = np.random.random(5)
print("\nArray of randoms:", arr4)
print("Elements type:", arr4.dtype)

# Array of Normal Random Variables
arr5 = np.random.randn(5)
print("\nArray of standard normal variables:", arr5)
print("Elements type:", arr5.dtype)

# Array of Equally Spaced Elements
arr6 = np.linspace(-10, 10, 101) # pradinis, galutinis, kiek elementu reikia sugeneruoti (suskaido intervala i100 daliu)
print("\nArray of equally distributed values:", arr6)
print("Elements type:", arr6.dtype)

# Array by Range
arr7 = np.arange(-10, 10 + 0.2,0.2 ) # pradinis, atstumas tarp tasku turi buti 0,5
arr7 = np.around(arr7,2) # suapvalina skaicius iki  skaiciu po kablelio
print("\nArray of elements in range:", arr7)
print("Elements type:", arr7.dtype)
```

# 2. Line Chart and Scatter Plot

## 2.1 Linear Function Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 2)
y = 2 * x - 7

# Plot
plt.plot(x, y)

# Adding points to the plot (grafikinis grafikas scatter --> taskai)
plt.scatter(x, y, color="red", s=25)

# Add center point: To Do
xc = (x[0] + x[1]) / 2
print("xc: " , xc)
yc = (y[0] + y[1]) / 2
print("yc: ", yc)
plt.scatter(xc, yc, color="lightgreen", s=25)



# Plot title (Grafiko pavadinimas)
plt.title("\nLinear Function: y=2*x - 7\n", fontsize=14, weight="bold", color="red")

# Axis Labels (Asiu pavadinimai)
plt.xlabel("X", fontsize = 10, weight="bold", color="red")
plt.ylabel("Y", fontsize = 10, weight="bold", color="red")
# Grid
plt.grid()

# Display plot
plt.show()
```

## 2.3 Dependend Data Versus Independent Data

### 2.3.1 Generating Independent Data

```python
# Number of data points
N = 100

# Vector X
vx2 = np.linspace(-10, 10, N)

# Vector Y does not depend on X
vy2 = np.random.randn(100)
```

### 2.3.2 Generating Linearly Dependent Data

```python
# Number of points
N = 100

# Vector X
vx3 = np.linspace(-10, 10, N)

# Vector Y
vy3 = [2*x + 8 + 3*np.random.randn(1) for x in vx2]

# Fitting Linear Regression vy = f(vx)
a, b = np.polyfit(vx3, vy3, 1)

vfit = a * vx3 + b
```

# 3. Histograms, Frequency and Density Plots

## 3.1 What distribution np.random.random() relies upon?

```python
import numpy as np
import matplotlib.pyplot as plt

# Data [0,1 ] intervalas
random_points = np.random.random(1000)
# print(random_points)

# Sample mean
sample_mean = np.mean(random_points)
print("Sample Mean:", sample_mean)

# Sample standard deviation (paklaida)
sample_std = np.std(random_points)
print("Sample Standar Deviation:", sample_std)

# Sample variance (vidurkis)
sample_variance = np.var(random_points)
print("Sample Variance:", sample_variance)

# Plotting space (uzduoda bendra erdve(lapa) grafikams, wspace, hspace yra atstumas tarp grafiku)
plt.figure(figsize=(11, 5))
plt.subplots_adjust(wspace=0.25, hspace=0.25)

# Histogram colors (grafiko ir srifto salvos)
color = "#4f93fe"
font_color = "#01577d"

# Frequency histogram (daznio grafikas)
plt.subplot(1, 2, 1)# braizomi 2 grafikai, sitas 1
plt.hist(random_points, bins=10, density=False, color=color, edgecolor="white")# braizomi 2 grafikai, sitas 2
# Frequency plot: title
plt.title("\nFrequency of np.random.random()\n", fontsize=14, weight="bold", color=font_color)
# Frequency plot: labels
plt.ylabel("Frequency", fontsize=10, color=font_color)
plt.xlabel("Value", fontsize=10, color=font_color)
# Frequency plot: axes tick symbols
plt.xticks(fontsize=10, color=font_color)
plt.yticks(fontsize=10, color=font_color)

# Density Histogram
plt.subplot(1, 2, 2)
plt.hist(random_points, bins=10, density=True, color=color, edgecolor="white")
# Density plot: title
plt.title("\nDensity of np.random.random()\n", fontsize=14, weight="bold", color=font_color)
# Density plot: labels
plt.ylabel("Density", fontsize=10, color=font_color)
plt.xlabel("Value", fontsize=10, color=font_color)
# Density plot: axes tick symbols
plt.xticks(fontsize=10, color=font_color)
plt.yticks(fontsize=10, color=font_color)

# Display Plot
plt.show()
```

**Answer:** np.random.random() takes values from uniform distribution on [0, 1]

## 3.2. Standard Normal Distribution

```python
import numpy as np

# Statndard Normal Distribution Function

def normal_df(x, mu=0, sigma=1): # mu vidurkis
    return 1 / sigma / np.sqrt(2*np.pi) * np.exp(-1/2 * ((x - mu)/sigma)**2) # naudojama nubraizyti tankiui

vectorized_normal_df = np.vectorize(normal_df) # vektorizacija, funkscija pritaikyti kiekvienam dydziui

x = np.linspace(-3.5, 3.5, 100) # sugeneruoja x, 100 tasku, [ nd -> [5, 35]]
y = vectorized_normal_df(x)

print("x:", x)
print("y:", y)

plt.plot(x,y) 

# Sample from Standard Norma Distribution
stnorm_points = np.random.normal(loc=0, scale=1, size=1000) #  nd-> loc 20, scale 5
plt.hist(stnorm_points, bins=15, density=True, label="sample histogram") #label naudojmas legendoj
plt.title("\nStandard Normal Distribution Function \nvs\n Sample Histogram\n")# nd pakeisti pavadinim

# Add Standard Normal Distribution Density Function
plt.plot(x, y, label="standard normal density") # nd rasyti density

# Add grid
plt.grid()

# Add legend
plt.legend()

# Display Plot
plt.show() 
```
